# Посчитали и добавили в таблицу
цену квадратного метра;
день недели, месяц и год публикации объявления;
этаж квартиры; варианты — первый, последний, другой;
соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
df['price_metr']=df['last_price']/df['total_area']
df['price_metr']=df['price_metr'].astype(int)
df
```

```python
df['year']=pd.DatetimeIndex(df['first_day_exposition']).year
df['month']=pd.DatetimeIndex(df['first_day_exposition']).month
df['day']=pd.DatetimeIndex(df['first_day_exposition']).day
df
```

```python
ft=df['floors_total'].median()
df['floors_total']=df['floors_total'].fillna(ft)
df['floors_total']=df['floors_total'].astype(int)
def floor_category(column):
    if column[8]==1:
        return "Первый"
    elif column[8]==column[6]:
        return "Последний"
    else:
        return "Другой"
df['kakoi_etazh'] = df.apply(floor_category, axis = 1)
df
```

```python
la=df['living_area'].median()
df['living_area']=df['living_area'].fillna(la)
```

```python
df['sootlata']=df['living_area']/df['total_area']
```

```python
df['sootkata']=df['kitchen_area']/df['total_area']
df
```

# Изучили время продажи квартиры. Построили гистограмму. Посчитали среднее и медиану. 

```python
X = df['total_area']
Y = df['price_metr']
fig, ax = plt.subplots(figsize=(9, 9))

plt.bar(X,Y, color="purple",width=5)
plt.title('Площадь и цена за кв')
plt.xlabel('Площадь')
plt.ylabel('Цена за кв')
plt.grid()
plt.show()
```

```python
df.describe()
```

```python
r=df['rooms'].median()
df['rooms']=df['rooms'].fillna(r)
df['rooms']=df['rooms'].replace(0,1)
df['rooms']=df['rooms'].astype(int)
```

```python

C = df['rooms']
D = df['last_price']
fig, ax = plt.subplots(figsize=(9, 9))

plt.bar(C,D, color="green",width=.5)
plt.title('Комнаты и цена')
plt.xlabel('комнаты')
plt.ylabel('Цена квартиры')
plt.xlim(0,20)
plt.grid()
plt.show()
```

```python
ch=df['ceiling_height'].median()
df['ceiling_height']=df['ceiling_height'].fillna(ch)
F = df['ceiling_height']
D = df['last_price']
fig, ax = plt.subplots(figsize=(9, 9))

plt.bar(F,D, color="green",width=3)
plt.title('Высота потолка и цена')
plt.xlabel('Высота потолка')
plt.ylabel('Цена квартиры')
plt.grid()
plt.show()
```

# Изучили время продажи квартиры. Построили гистограмму. 

```python
de=df['days_exposition'].median()
df['days_exposition']=df['days_exposition'].fillna(de)
df['days_exposition']=df['days_exposition'].astype(int)
G = df['days_exposition']
H = df['first_day_exposition']
fig, ax = plt.subplots(figsize=(9, 9))

plt.bar(H,G, color="red",width=3)
plt.title('Дата размещения и продажи')
plt.xlabel('дата размещения')
plt.ylabel('дни в продаже')
plt.grid()
plt.show()
```
